```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from google.colab import files
uploaded = files.upload()
# Use the filename dynamically
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename, names=column)
df.head(150)
```

```python
df.describe()
```

```python
sns.pairplot(df,hue='class_labels')
```

```python
#seperate features and target
data=df.values
X = data[:,:4]
y = data[:,4]
print(X)
```

splittling the data into training and testing

```python
#split the data into train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train)
```

model 1: support vector machine algorithm

```python
#support vector mchine algorithm
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction = model_svc.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction))
```

model 2:logistic regression

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
```

```python
prediction =model_lr.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction)*100)
```

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train, y_train)
```

```python
prediction =model_dt.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction))
```

```python
#detailed classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
```
